# Handeling missing values

```python
data.isnull().sum()
```

```python
#since 327 data values out of 418 of column cabin is missing, we can delete that column completely.
del data["Cabin"]
```

```python
data.columns
```

```python
#handling missing values for age ---> replacing with median
data["Age"].fillna(data["Age"].median(),inplace=True)
```

```python
data.isna().sum()
```

```python
#handeling missing value for fare ---> replacing with mean
data["Fare"].fillna(data["Fare"].median(),inplace=True)
```

```python
data.isna().sum()
```

```python
#Hence all the missing values havebeen handled
```

# Seperating dependent and independent values

```python
column = ['Name','Ticket','Fare']
data = data.drop(column,axis=1)
```

```python
data
```

```python
x = data.drop('Survived',axis=1)
y = data[['Survived']]
```

# encode categorical columns

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
x["Sex"] = le.fit_transform(x["Sex"])
x["Embarked"] = le.fit_transform(x["Embarked"])
x
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x = sc.fit_transform(x)
x
```
